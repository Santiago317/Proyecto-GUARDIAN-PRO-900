# TRANSFORM - Transformar los datos que necesitamos

```python
df3_union = pd.concat([df_drive, df_editar,],sort=True)
#definimos la union de los dataframe con uno nuevo para poder utilizar esa informacion
```

```python
df3_union.describe
#de este modo podemos apreciar la informacion unida de los dos dataframe.
```

```python
df3_union.dropna(axis =1)
#los datos anteriores cuentan con valores nulos, lo que hicimos fue eliminar esa columna con esos valores.
```

```python
df3_union. dtypes
#observamos que son de tipo object los datos
```

```python
df3_union ['Name'] = df3_union ['Unnamed: 0']. astype (float)
df3_union ['Date'] = df3_union ['Unnamed: 0']. astype (float)

#convertimos la columna Name y Date en flotante
```

```python
df3_union. dtypes
#podemos observar que ahora son flotantes
```

# LOAD - Cargar los datos que tenemos


```python
from matplotlib import pyplot as plt 
#llama a la libreria de matplot
%matplotlib inline  
#plotea en linea
import seaborn as sns 
#libreria para realizar graficas de manera matricial y detectar patrones
```

```python
ploteo_nombre = df3_union[["Name"]]
ploteo_datos = df3_union[["Date"]]
#plotea numeros, no grafica texto
```

```python
x = (ploteo_nombre)
y = (ploteo_datos)

plt.plot(x,y)
plt.xlabel('cantidad de Nombres ingresados RFDI')
plt.ylabel('Cantidad de Datos que accedieron')
plt.title('Registros RFID')
plt.show()

#en esta grafica podemos ver la cantidad de nombres que ingresan al cenote
#va incrementando, con forme la gente se va registrando.
```

```python
# Modulos con los que trabajaremos para crear la grafica pero 3D
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

# Creamos la figura
fig = plt.figure()

# Agrrgamos un plano 3D
ax1 = fig.add_subplot(111,projection='3d')

# Datos en array bi-dimensional
x = (ploteo_nombre)
z = (ploteo_datos)


# plot_wireframe nos permite agregar los datos x, y, z. Por ello 3D
# Es necesario que los datos esten contenidos en un array bi-dimensional
ax1.plot_wireframe(x, y, z)

# Mostramos el gr√°fico
plt.show()
```
