```python
df3_union = pd.concat([df_drive, df_editar,],sort=True)
#definimos la union de los dataframe con uno nuevo para poder utilizar esa informacion
```

```python
df3_union.describe
#de este modo podemos apreciar la informacion unida de los dos dataframe.
```

```python
df3_union.dropna(axis =1)
#los datos anteriores cuentan con valores nulos, lo que hicimos fue eliminar esa columna con esos valores.
```
